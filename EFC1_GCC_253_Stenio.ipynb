##  Insertion Sort


```python
def insertionSort(arranjo, numComparacao):
  for j in range(1, len(arranjo)):
    numComparacao+=1 
    chave = arranjo[j]
    i=j-1
    while i>=0 and arranjo[i]>chave:
      numComparacao+=1 
      arranjo[i+1]=arranjo[i]
      i=i-1 
    arranjo[i+1]=chave
  
  return arranjo, numComparacao

# grafico insertionSort para casos aleatorios 
aux = []
for i in range (1, 100):  
  A = geraArranjoAleatorio(i)
  aux.append(tempo(insertionSort, A))
  
plt.plot(aux, 'b', label='insertionSort')
plt.legend()
plt.show() 


```

##  Selection Sort


```python
def selectionSort(arranjo, numComparacao):
  for i in range(len(arranjo)):
    numComparacao+1 
    menor_i = i

    for j in range(i+1, len(arranjo)):      
      numComparacao+=1 
      if arranjo[menor_i] > arranjo[j]:
        numComparacao+=1 
        menor_i = j

    numComparacao+=1 
    arranjo[i], arranjo[menor_i] = arranjo[menor_i], arranjo[i]
  
  numComparacao+=1 
  return arranjo, numComparacao

# selectionSort Casos Aleatorios 
aux = [] 
for i in range (0, 100): 
  A = geraArranjoAleatorio(i)
  aux.append(tempo(selectionSort, A))

plt.plot(aux, 'b', label='selectionSort')
plt.legend()
plt.show() 
```

##  Bubble Sort


```python
def bubbleSort(arranjo, numComparacao):
  for i in range(len(arranjo)):
    numComparacao+=1  
    for j in range(len(arranjo) - 1): 
      numComparacao+=1 
      if (arranjo[j] > arranjo[j+1]):
        numComparacao+=1  
        temp = arranjo[j+1]
        arranjo[j+1] = arranjo[j] 
        arranjo[j] = temp
      numComparacao+=1
    numComparacao+=1 
  numComparacao+=1 

  return arranjo, numComparacao

# bubbleSort Casos Aleatorios
aux = [] 
for i in range (0, 100): 
  A = geraArranjoAleatorio(i)
  aux.append(tempo(bubbleSort, A))

plt.plot(aux, 'b', label='bubbleSort')
plt.legend()
plt.show() 
```

## Heap Sort

```python
def heapify(arr, n, i, numComparacao):
  largest = i  
  l = 2 * i + 1     
  r = 2 * i + 2     

  if l < n and arr[largest] < arr[l]:
    numComparacao+=1
    largest = l
    
  if r < n and arr[largest] < arr[r]:
    numComparacao+=1
    largest = r

  if largest != i:
    numComparacao+=1 
    arr[i], arr[largest] = arr[largest], arr[i] 
    heapify(arr, n, largest)
  
  
def heapSort(arr, numComparacao):
  n = len(arr)

  for i in range(n//2 - 1, -1, -1):
    numComparacao+=1
    heapify(arr, n, i)
  
  for i in range(n-1, 0, -1):
    numComparacao+=1
    arr[i], arr[0] = arr[0], arr[i]  
    heapify(arr, i, 0)

  return numComparacao

# heapSort Casos Aleatorios
aux = [] 

for i in range (0, 10): 
  A = geraArranjoAleatorio(i)
  aux.append(tempo(selectionSort, A))

plt.plot(aux, 'b', label='heapSort')
plt.legend()
plt.show() 
```

## Quick Sort

> pivo = *A[A-comprimento]*

```python
def particao(arranjo,p,r, comparacao):
  pivo = arranjo[r]
  i = p 
  for j in range(p, r):
    comparacao+=1 
    if arranjo[j]<=pivo:
      comparacao+=1 
      aux = arranjo[i]
      arranjo[i] = arranjo[j]
      arranjo[j] = aux
      i+=1
      
  comparacao+=1 
  aux = arranjo[i]
  arranjo[i] = arranjo[r]
  arranjo[r] = aux
  return i, comparacao

def quickSort(arranjo,p,r, comparacao):
  ini= time.time()
  if p < r:
    comparacao+=1 
    q, comparacao = particao(arranjo,p,r, comparacao)
    quickSort(arranjo,p,q-1, comparacao)
    quickSort(arranjo,q+1,r, comparacao)
  comparacao+=1 
  fim = time.time()
  tempo_exec=fim-ini
  print(comparacao)
  return tempo_exec


# quickSort Casos aleatorios
aux = []
comparacao = 0
for i in range (0, 100): 
  A = geraArranjoAleatorio(i)
  aux.append(quickSort(A,0, len(A)-1, comparacao))

plt.plot(aux, 'b', label='quickSort')
plt.legend()
plt.show() 

```
